```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv(r"Dataset\Titanic-Dataset.csv")
data
```

```python
data.isnull().sum()
```

```python
data["Age"].fillna(value = data["Age"].mean(), inplace = True)
data["Cabin"].fillna(value = "No_Cabin", inplace = True)
data["Embarked"].fillna(value = "N", inplace = True)
```

```python
data.isnull().sum()
```

```python
sns.histplot(data = data["Sex"], kde = True)
```

```python
sns.histplot(data = data["Fare"], kde = True)
```

```python
sns.histplot(data = data["Age"], kde = True)
```

```python
sns.boxplot(data = data, x = "Sex", y = "Age", hue = "Survived")
```

```python
sns.boxplot(data = data, x = "Sex", y = "Survived")
```

```python
sns.boxplot(data = data, x = "Sex", y = "Fare", hue = "Survived")
```
